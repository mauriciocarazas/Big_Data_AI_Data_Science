```python
# Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargamos la base de datos
df = pd.read_csv("/content/drive/MyDrive/DATASETS/covid_worldwide.csv")

# Indicamos la cantidad de filas y columnas
filas, columnas = df.shape
print(f"La base de datos tiene {filas} filas y {columnas} columnas")

print("Visualización de la base de datos:")
print(df.head())

# Describimos la base de datos
descripcion = df.describe()
print(descripcion)

# Seleccionamos solo las columnas numéricas
df_numeric = df.select_dtypes(include=[np.number])

# Realizamos una revisión estadística
correlacion = df_numeric.corr()
sns.heatmap(correlacion, annot=True)
plt.title("Correlación entre variables numéricas")
plt.show()

# Generamos tres gráficas de dispersión
sns.scatterplot(data=df, x="Country", y="Total Cases")
plt.title("Gráfica 1")
plt.show()

sns.scatterplot(data=df, x="Country", y="Total Deaths")
plt.title("Gráfica 2")
plt.show()

sns.scatterplot(data=df, x="Total Cases", y="Total Deaths")
plt.title("Gráfica 3")
plt.show()
```

```python
  import numpy as np
  import pandas as pd
  import matplotlib.pyplot as plt
  import seaborn as sns
  from pandas_datareader import data as pdr
  import yfinance as yf
  from sklearn.preprocessing import StandardScaler
  from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
  from sklearn.linear_model import LinearRegression
  from sklearn.tree import DecisionTreeRegressor
  from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
  from sklearn.svm import SVR
  from sklearn.neural_network import MLPRegressor
  from sklearn.neighbors import KNeighborsRegressor


  # Descargar los datos de Yahoo Finance
  yf.pdr_override()

  # Definir el rango de fechas de inter√©s
  start_date = "2015-01-01"
  end_date = "2023-03-31"



  start_date = "2015-01-01"
  end_date = "2023-03-31"

  aapl = pdr.get_data_yahoo('AAPL', start=start_date, end=end_date)['Adj Close']


  msft = pdr.get_data_yahoo('MSFT', start=start_date, end=end_date)['Adj Close']


  goog = pdr.get_data_yahoo('GOOG', start=start_date, end=end_date)['Adj Close']

  fb = pdr.get_data_yahoo('META', start=start_date, end=end_date)['Adj Close']


  amzn = pdr.get_data_yahoo('AMZN', start=start_date, end=end_date)['Adj Close']


  sp500 = yf.download('^GSPC', start=start_date, end=end_date)['Adj Close']

  dowjones = yf.download('^DJI', start=start_date, end=end_date)['Adj Close']



  # Separar los datos en entrenamiento y prueba
  train_size = 0.9
  train_end = int(len(aapl) * train_size)

  aapl_train = aapl[:train_end]
  msft_train = msft[:train_end]
  goog_train = goog[:train_end]
  fb_train = fb[:train_end]
  amzn_train = amzn[:train_end]
  sp500_train = sp500[:train_end]
  dowjones_train = dowjones[:train_end]

  aapl_test = aapl[train_end:]
  msft_test = msft[train_end:]
  goog_test = goog[train_end:]
  fb_test = fb[train_end:]
  amzn_test = amzn[train_end:]
  sp500_test = sp500[train_end:]
  dowjones_test = dowjones[train_end:]


  # Tambi√©n es necesario definir un time delay o lag correspondiente
  lag = 7


  # Normalizar los datos para que tengan media cero y desviaci√≥n est√°ndar uno
  scaler = StandardScaler()
  aapl_train_norm = scaler.fit_transform(aapl_train.to_numpy().reshape(-1, 1)).flatten()

  msft_train_norm = scaler.transform(msft_train.to_numpy().reshape(-1, 1)).flatten()
  goog_train_norm = scaler.transform(goog_train.to_numpy().reshape(-1, 1)).flatten()
  fb_train_norm = scaler.transform(fb_train.to_numpy().reshape(-1, 1)).flatten()
  amzn_train_norm = scaler.transform(amzn_train.to_numpy().reshape(-1, 1)).flatten()
  sp500_train_norm = scaler.transform(sp500_train.to_numpy().reshape(-1, 1)).flatten()
  dowjones_train_norm = scaler.transform(dowjones_train.to_numpy().reshape(-1, 1)).flatten()

  aapl_test_norm = scaler.transform(aapl_test.to_numpy().reshape(-1, 1)).flatten()
  msft_test_norm = scaler.transform(msft_test.to_numpy().reshape(-1, 1)).flatten()
  goog_test_norm = scaler.transform(goog_test.to_numpy().reshape(-1, 1)).flatten()
  fb_test_norm = scaler.transform(fb_test.to_numpy().reshape(-1, 1)).flatten()
  amzn_test_norm = scaler.transform(amzn_test.to_numpy().reshape(-1, 1)).flatten()
  sp500_test_norm = scaler.transform(sp500_train.to_numpy().reshape(-1, 1)).flatten()
  dowjones_test_norm = scaler.transform(dowjones_train.to_numpy().reshape(-1, 1)).flatten()


  # Crear las matrices de caracter√≠sticas y etiquetas
  def create_features_and_labels(data, lag):
      X = []
      y = []
      for i in range(lag, len(data)):
          X.append([data[i-lag:i]])
          y.append(data[i])
      X = np.array(X)
      y = np.array(y)
      X = X.reshape(X.shape[0], X.shape[2])
      return X, y

  aapl_train_X, aapl_train_y = create_features_and_labels(aapl_train_norm, lag)
  msft_train_X, msft_train_y = create_features_and_labels(msft_train_norm, lag)
  goog_train_X, goog_train_y = create_features_and_labels(goog_train_norm, lag)
  fb_train_X, fb_train_y = create_features_and_labels(fb_train_norm, lag)
  amzn_train_X, amzn_train_y = create_features_and_labels(amzn_train_norm, lag)
  sp500_train_X, sp500_train_y = create_features_and_labels(sp500_train_norm, lag)
  dowjones_train_X, dowjones_train_y = create_features_and_labels(dowjones_train_norm, lag)

  aapl_test_X, aapl_test_y = create_features_and_labels(aapl_test_norm, lag)
  msft_test_X, msft_test_y = create_features_and_labels(msft_test_norm, lag)
  goog_test_X, goog_test_y = create_features_and_labels(goog_test_norm, lag)
  fb_test_X, fb_test_y = create_features_and_labels(fb_test_norm, lag)
  amzn_test_X, amzn_test_y = create_features_and_labels(amzn_test_norm, lag)
  sp500_test_X, sp500_test_y = create_features_and_labels(sp500_test_norm, lag)
  dowjones_test_X, dowjones_test_y = create_features_and_labels(dowjones_test_norm, lag)

  max_len = 201

  # Redimensionar los arrays de test para que tengan la misma longitud
  aapl_test_X = aapl_test_X[-max_len:]
  msft_test_X = msft_test_X[-max_len:]
  goog_test_X = goog_test_X[-max_len:]
  fb_test_X = fb_test_X[-max_len:]
  amzn_test_X = amzn_test_X[-max_len:]
  sp500_test_X = sp500_test_X[-max_len:]
  dowjones_test_X = dowjones_test_X[-max_len:]

# Crear la matriz de caracter√≠sticas final
X_test = np.column_stack((aapl_test_X, msft_test_X, goog_test_X, fb_test_X, amzn_test_X, sp500_test_X, dowjones_test_X))

# Estandarizar los datos
X_test = X_test.reshape(X_test.shape[0], -1)

X_test = scaler.fit_transform(X_test)




  # Crear la matriz de caracter√≠sticas final
  X_train = np.column_stack((aapl_train_X, msft_train_X, goog_train_X, fb_train_X, amzn_train_X, sp500_train_X, dowjones_train_X))
  X_test = np.column_stack((aapl_test_X, msft_test_X, goog_test_X, fb_test_X, amzn_test_X, sp500_test_X, dowjones_test_X))
  #print(X_train.head())
  #print(X_test.head())
  # Crear la etiqueta final
  y_train = aapl_train_y
  y_test = aapl_test_y
 # print(y_train.head())
  #print(y_test.head())

  
  X_train = np.nan_to_num(X_train, nan=0)

  X_test = np.nan_to_num(X_test, nan=0)

  y_train = np.nan_to_num(y_train, nan=0)

  y_test = np.nan_to_num(y_test, nan=0)

  # Implementar la regresi√≥n lineal multivariada con gradiente descendiente
class MultivariateLinearRegression:
    def __init__(self, lr=0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def train(self, X_train, y_train):
        n_samples, n_features = X_train.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            y_pred = np.dot(X_train, self.weights) + self.bias
            dw = (1 / n_samples) * np.dot(X_train.T, (y_pred - y_train))
            db = (1 / n_samples) * np.sum(y_pred - y_train)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        y_pred = np.dot(X, self.weights) + self.bias
        return y_pred





  def evaluate_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"MSE: {mse:.3f}")
    print(f"RMSE: {rmse:.3f}")
    print(f"MAE: {mae:.3f}")
    print(f"R^2: {r2:.3f}")


  # Entrenar y evaluar diferentes algoritmos de machine learning
  lr = MultivariateLinearRegression()
  lr.train(X_train, y_train)
  lr_pred_train = lr.predict(X_train)
  lr_pred_test = lr.predict(X_test)

  ##fill NAN
  lr_pred_train = np.nan_to_num(lr_pred_train, nan=0)
  lr_pred_test = np.nan_to_num(lr_pred_test, nan=0)

  rf = RandomForestRegressor(random_state=42)
  rf.fit(X_train, y_train)
  rf_pred_train = rf.predict(X_train)
  rf_pred_test = rf.predict(X_test)

  gb = GradientBoostingRegressor(random_state=42)
  gb.fit(X_train, y_train)
  gb_pred_train = gb.predict(X_train)
  gb_pred_test = gb.predict(X_test)

  svr = SVR()
  svr.fit(X_train, y_train)
  svr_pred_train = svr.predict(X_train)
  svr_pred_test = svr.predict(X_test)

  knn = KNeighborsRegressor()
  knn.fit(X_train, y_train)
  knn_pred_train = knn.predict(X_train)
  knn_pred_test = knn.predict(X_test)




  #Evaluar desempe√±o de los algoritmos
  print("Metricas de desempeno para los diferentes algoritmos:\n")

  print("Multivariate Linear Regression:")
  print("Training Set:")
  evaluate_metrics(y_train, lr_pred_train)
  print("\nTest Set:")
  evaluate_metrics(y_test, lr_pred_test)

  print("\nRandom Forest Regressor:")
  print("Training Set:")
  evaluate_metrics(y_train, rf_pred_train)
  print("\nTest Set:")
  evaluate_metrics(y_test, rf_pred_test)

  print("\nGradient Boosting Regressor:")
  print("Training Set:")
  evaluate_metrics(y_train, gb_pred_train)
  print("\nTest Set:")
  evaluate_metrics(y_test, gb_pred_test)

  print("\nSupport Vector Regressor:")
  print("Training Set:")
  evaluate_metrics(y_train, svr_pred_train)
  print("\nTest Set:")
  evaluate_metrics(y_test, svr_pred_test)

  print("\nK Nearest Neighbors Regressor:")
  print("Training Set:")
  evaluate_metrics(y_train, knn_pred_train)
  print("\nTest Set:")
  evaluate_metrics(y_test, knn_pred_test)

  #Visualizar predicciones vs valores reales
  plt.figure(figsize=(15,5))
  plt.plot(y_test, label="Real")
  plt.plot(lr_pred_test, label="Multivariate Linear Regression")
  plt.plot(rf_pred_test, label="Random Forest Regressor")
  plt.plot(gb_pred_test, label="Gradient Boosting Regressor")
  plt.plot(svr_pred_test, label="Support Vector Regressor")
  plt.plot(knn_pred_test, label="K Nearest Neighbors Regressor")
  plt.legend()
  plt.title("Predicciones vs valores reales (Test Set)")
  plt.show()




  #Selecci√≥n del mejor algoritmo seg√∫n la m√©trica de desempe√±o
  def select_best_algorithm(y_test, lr_pred_test, rf_pred_test, gb_pred_test, svr_pred_test, knn_pred_test):
    """
    Esta funci√≥n compara las m√©tricas de los diferentes algoritmos y devuelve el mejor algoritmo y la mejor m√©trica.
    Par√°metros:
        - y_test: array con los valores reales de y en el conjunto de prueba.
        - lr_pred_test: array con las predicciones del modelo de regresi√≥n lineal en el conjunto de prueba.
        - rf_pred_test: array con las predicciones del modelo de random forest en el conjunto de prueba.
        - gb_pred_test: array con las predicciones del modelo de gradient boosting en el conjunto de prueba.
        - svr_pred_test: array con las predicciones del modelo de support vector regression en el conjunto de prueba.
        - knn_pred_test: array con las predicciones del modelo de k-nearest neighbors en el conjunto de prueba.
    Retorna:
        - best_algorithm: El mejor algoritmo entre los evaluados.
        - best_metric: La mejor m√©trica obtenida entre los modelos evaluados.
    """
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

    # C√°lculo de las m√©tricas para cada modelo
    lr_r2 = r2_score(y_test, lr_pred_test)
    lr_mae = mean_absolute_error(y_test, lr_pred_test)
    lr_mse = mean_squared_error(y_test, lr_pred_test)

    rf_r2 = r2_score(y_test, rf_pred_test)
    rf_mae = mean_absolute_error(y_test, rf_pred_test)
    rf_mse = mean_squared_error(y_test, rf_pred_test)

    gb_r2 = r2_score(y_test, gb_pred_test)
    gb_mae = mean_absolute_error(y_test, gb_pred_test)
    gb_mse = mean_squared_error(y_test, gb_pred_test)

    svr_r2 = r2_score(y_test, svr_pred_test)
    svr_mae = mean_absolute_error(y_test, svr_pred_test)
    svr_mse = mean_squared_error(y_test, svr_pred_test)

    knn_r2 = r2_score(y_test, knn_pred_test)
    knn_mae = mean_absolute_error(y_test, knn_pred_test)
    knn_mse = mean_squared_error(y_test, knn_pred_test)

    # Selecci√≥n del mejor algoritmo y la mejor m√©trica
    algorithms = ['Linear Regression', 'Random Forest', 'Gradient Boosting', 'Support Vector Regression', 'K-Nearest Neighbors']
    r2_scores = [lr_r2, rf_r2, gb_r2, svr_r2, knn_r2]
    mae_scores = [lr_mae, rf_mae, gb_mae, svr_mae, knn_mae]
    mse_scores = [lr_mse, rf_mse, gb_mse, svr_mse, knn_mse]

    best_r2 = max(r2_scores)
    best_r2_index = r2_scores.index(best_r2)
    best_r2_algorithm = algorithms[best_r2_index]
    best_r2_metric = 'R^2'

    best_mae = min(mae_scores)
    best_mae_index = mae_scores.index(best_mae)
    best_mae_algorithm = algorithms[best_mae_index]
    best_mae_metric = 'MAE'

    if best_r2 >= best_mae:
        return best_r2_algorithm, best_r2_metric
    else:
        return best_mae_algorithm, best_mae_metric

  best_algorithm, best_metric = select_best_algorithm(y_test, lr_pred_test, rf_pred_test, gb_pred_test, svr_pred_test, knn_pred_test)
  print("\nLa mejor m√©trica de desempe√±o es: ", best_metric, " El mejor algoritmo seg√∫n la m√©trica de desempe√±o:", best_algorithm)
```

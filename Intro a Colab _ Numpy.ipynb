```python
import numpy

n = 8
m = 8
X = numpy.random.randint(0, 255, (n, m))

Y = numpy.array([[0, 1, 0],
                 [1, -4, 1],
                 [0, 1, 0]])

Z = numpy.zeros_like(X)

for i in range(1, n-1):
    for j in range(1, m-1):
        Z[i,j] = numpy.sum(X[i-1:i+2,j-1:j+2] * Y)

        if Z[i,j] > 255:
            Z[i,j] = 255
     
        elif Z[i,j] < 0:
            Z[i,j] = 0



Z[0,: ] = X[0,:]
Z[-1,:] = X[-1,:]
Z[:,0 ] = X[:,0]
Z[:,-1] = X[:,-1]  



print(Z)
```
